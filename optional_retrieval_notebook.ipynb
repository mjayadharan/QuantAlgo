```python
import requests
import pandas as pd
import os
```

```python
from datetime import datetime, timedelta

def get_all_dates_in_year(year):
    # Start from the first day of the year
    start_date = datetime(year, 1, 1)
    # Calculate the number of days in the year
    days_in_year = 366 if (start_date.year % 4 == 0 and (start_date.year % 100 != 0 or start_date.year % 400 == 0)) else 365
    
    # Generate a list of all dates in the year
    dates = [(start_date + timedelta(days=i)).strftime('%Y-%m-%d') for i in range(days_in_year)]
    
    return dates

# Example usage:
# year = 2024
# all_dates = get_all_dates_in_year(year)
# for date in all_dates:
#     print(date)

def create_folder_if_not_exists(folder_name):
    # Check if the folder exists
    if not os.path.exists(folder_name):
        # Create the folder if it doesn't exist
        os.makedirs(folder_name)
        print(f"Folder '{folder_name}' created.")
    else:
        print(f"Folder '{folder_name}' already exists.")

# # Example usage:
# folder_name = "my_new_folder"
# create_folder_if_not_exists(folder_name)

def get_option_data(date, symbol, api_key, data_type="json"):
    url = 'https://www.alphavantage.co/query?function={}&symbol={}&date={}&apikey={}&datatype={}'.format(function,
                                                                                             symbol,
                                                                                             date,
                                                                                             api_key,
                                                                                             data_type)

    r = requests.get(url)
    data = r.json()
    try:
        return pd.DataFrame(data['data'])
    except Exception as e:
        print("**Warning: {}".format(e))
        return pd.DataFrame()
```

```python
api_key = "7OVPSPQGG82Z6EJZ"


for symbol in symbol_list:
    for year in year_list:
        date_list = get_all_dates_in_year(year)
        

        try:
            create_folder_if_not_exists("option_data")
            create_folder_if_not_exists(os.path.join("option_data",symbol))
            create_folder_if_not_exists(os.path.join("option_data",symbol, str(year)))
            if os.path.exists(os.path.join("option_data",symbol, str(year), "option_data_{}.csv".format(str(year)))):
                print("Skipping because data already exists for {}: {}".format(symbol, year))
                
                continue

            df_list = [get_option_data(date, symbol, api_key) for date in date_list]
            year_op_data = pd.concat(df_list)
            year_op_data.to_csv(os.path.join("option_data",symbol, str(year), "option_data_{}.csv".format(str(year))))

            print("Done saving data for {}: {}".format(symbol, year))
        except Exception as e:
            print("**Warning: Error occured during data retrieval for {} for year {} : ".format(symbol, year,
                                                                                                e))


# 
```
